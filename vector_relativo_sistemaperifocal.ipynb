```python
import pymcel as pc
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
p = 1
e = 0.7

tetaers = np.linspace(0, 2 * np.pi, 1000)
rs = p / (1 + e * np.cos(tetaers))

xfs = rs * np.cos(tetaers)
yfs = rs * np.sin(tetaers)

plt.plot(xfs, yfs)
plt.axis('equal')
plt.grid(True)
plt.tight_layout
plt.show()
```

```python
@interact(e=(0, 2, 0.1))
def dibuja_trayectoria(e):
    p = 1
    q = p 
    Q = p 

    tetaers = np.linspace(0, 2 * np.pi, 1000)
    rs = p / (1 + e * np.cos(tetaers))

    xfs = rs * np.cos(tetaers)
    yfs = rs * np.sin(tetaers)

    plt.figure(figsize=(6, 6))

    plt.plot(xfs, yfs)
    plt.xlim(-2* Q, 2* Q)
    plt.ylim(-2 * Q, 2 * Q)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
p = 1
e = 0.7

tetaers = np.linspace(0, 2 * np.pi, 1000)
rs = p / (1 + e * np.cos(tetaers))

xfs = rs * np.cos(tetaers)
yfs = rs * np.sin(tetaers)

plt.plot(xfs, yfs)
plt.axis('equal')
plt.grid(True)
plt.tight_layout
plt.show()
```

```python
m1 = 1
m2 = 1
M = m1 + m2

rvecs = np.array([xfs, yfs])
rvecs = rvecs.T

r1vecs = m2 / M * rvecs
r2vecs = - m1 / M * rvecs

#plt.plot(xfs, yfs, label='Trayectoria del sistema')
plt.plot(r1vecs[:,0], r1vecs[:,1], label='Trayectoria de m1')
plt.plot(r2vecs[:,0], r2vecs[:,1], label='Trayectoria de m2')
plt.axis('equal')                   
plt.grid(True)
plt.legend()        
plt.show()
```
