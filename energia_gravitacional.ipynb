## Sistema de dos cuerpos

```python
G = 1
```

```python
sistema = [
    dict(m=1,r=np.array([0, 0, 0]),v=np.array([0, 0, 0])),
    # dict(m=0.01,r=np.array([1, 0, 0]),v=np.array([0, 3, 0])) se escapa
    # dict(m=0.01,r=np.array([1, 0, 0]),v=np.array([0, 1, 0])) no se escapa
    dict(m=0.01,r=np.array([1, 0, 0]),v=np.array([0, np.sqrt(2*G*1/1), 0])) #apenas se escapa
]
```

Energía total: $E=K+U$

```python
K = 0.5 * sistema[0]['m'] * np.linalg.norm(sistema[0]['v'])**2 + 0.5 * sistema[1]['m'] * np.linalg.norm(sistema[1]['v'])**2
```

```python
U = - G * sistema[0]['m'] * sistema[1]['m'] / np.linalg.norm(sistema[0]['r'] - sistema[1]['r'])
```

```python
E = K + U
print(E)
```

```python
ts = np.linspace(0, 10, 100)
rs, vs, rps, vps, cuadraturas = pc.ncuerpos_solucion(sistema, ts)
```

```python
plt.plot(rps[0,:,0], rps[0,:,1], label='Cuerpo 1')
plt.plot(rps[1,:,0], rps[1,:,1], label='Cuerpo 2')
plt.xlabel('x')                 
plt.ylabel('y')
plt.title('Trayectorias de los cuerpos')
plt.legend()
plt.axis('equal')
plt.grid()
plt.show()      
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=rps[0, :, 0], y=rps[0, :, 1], z=rps[0, :, 2],
    mode='lines',
    name='Cuerpo 1',
    line=dict(width=4)
))

fig.add_trace(go.Scatter3d(
    x=rps[1, :, 0], y=rps[1, :, 1], z=rps[1, :, 2],
    mode='lines',
    name='Cuerpo 2',
    line=dict(width=4)
))

fig.update_layout(
    title='Trayectorias de los cuerpos (3D)',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='z',
        aspectmode='data'
    ),
    legend=dict(x=0.01, y=0.99)
)

fig.show()
```
